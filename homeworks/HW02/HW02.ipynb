```python
plt.figure(figsize=(12,7))
plt.hist(df['age'], bins=15) # так как диапазон возрастов составляет около 40 лет, а объём выборки небольшой
plt.xlabel('Возраст')
plt.ylabel('Частота')
plt.title('Распределение возраста')
plt.savefig('figures/age_distribution.png')
plt.show()
```

```python
# Гистограмма отображает распределение возрастов пользователей. Значения на оси X - возраст, а высота столбцов показывает, сколько людей попадает
# в тот или иной возрастной диапазон.
# Выводы: основная масса пользователей находится в диапазоне примерно 20–60 лет - именно там сконцентрирована высокая частота.
# Видно наличие аномального значения около 120 лет, которое выбивается из общего распределения и требует проверки
```

```python
boxplot = df.boxplot(column=['revenue'], by='country', fontsize=12, rot=40, grid=True)
plt.ylabel('Revenue', fontsize=12)
plt.xlabel('Country', fontsize=12)
# Заголовок графика
plt.title('My boxplot: Revenue by Country', fontsize=11, fontweight='bold')

# Убираем автоматический заголовок от pandas (опционально)
plt.suptitle('')  # Удаляем суперзаголовок "Boxplot grouped by country"
plt.tight_layout()
# Показать график
plt.savefig('figures/boxplot_revenue_by_country.png')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
df.boxplot(column='revenue', by='country', ax=ax,
          fontsize=12, rot=45, grid=True)

# Подписи
ax.set_ylabel('Revenue', fontsize=12)
ax.set_xlabel('Country', fontsize=12)
ax.set_title('My Boxplot: Revenue by Country', 
             fontsize=14, fontweight='bold')

plt.suptitle('')
plt.tight_layout()

# Сохраняем
plt.savefig('figures/boxplot.png', 
            dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Боксплот сравнивает распределение выручки (revenue) между странами.
# Каждый «box» показывает медиану, квартили, разброс значений и возможные выбросы для каждой страны.
# Выводы: У DE наблюдается самый высокий разброс и самые большие значения revenue — есть клиенты, приносящие существенно больше остальных
# FR и US показывают более компактное распределение с меньшими медианами - это более ровные рынки без крупных клиентов.
# CN и RU находятся в среднем диапазоне, но у CN заметно большая вариативность по сравнению с RU.
```

```python
plt.figure(figsize=(10,6))
bins = [0, 18, 25, 35, 45, 55, 65, 80, 120]

plt.hist(df['age'], bins=bins, edgecolor='black')
plt.xlabel('Возраст')
plt.ylabel('Количество пользователей')
plt.title('Распределение возраста по возрастным группам')
plt.grid(axis='y', alpha=0.3)

plt.savefig('figures/age_distribution_v2.png')
plt.show()
```

```python
plt.figure(figsize=(10,6))

for country in df['country'].unique():
    subset = df[df['country'] == country]
    plt.scatter(subset['age'], subset['revenue'], label=country)

plt.xlabel('Возраст')
plt.ylabel('Выручка')
plt.title('Взаимосвязь между возрастом и выручкой')
plt.legend(title='Страна')
plt.grid(alpha=0.3)

plt.savefig('figures/age_vs_revenue_scatter.png')
plt.show()
```

```python
# Scatter plot используется именно тогда, когда нужно показать взаимосвязь между двумя количественными переменными
# Каждая точка - конкретный человек / наблюдение с двумя числовыми признаками: возрастом и объёмом выручки.
# Выводы: В правой части - человек примерно 120+ лет. Это невозможно, либо ошибка данных, либо это тестовая запись
# Нет явной зависимости вида: «чем старше - тем выше/ниже выручка». Точки разбросаны хаотично - связь слабая или отсутствует.
# Возрастная группа 20–50 лет - самая плотная. Большинство точек сосредоточено именно там - это нормальная рабочая выборка.
# Страна влияет на выручку заметно сильнее возраста.
```
